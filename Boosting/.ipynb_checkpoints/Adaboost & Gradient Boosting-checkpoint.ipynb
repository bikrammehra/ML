```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('../Data/mushrooms.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe().T
```

```python
sns.countplot(data=data, x='class')
```

```python
plt.figure(figsize=(8,4), dpi=200)
feature_unique = data.describe().T.reset_index().sort_values('unique')
sns.barplot(data=feature_unique, x='index', y='unique')
plt.xticks(rotation=90);
```

```python
X = pd.get_dummies(data.drop('class', axis=1), drop_first=True)
y = data['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=1)
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, plot_confusion_matrix, confusion_matrix, classification_report
```

```python
print(classification_report(y_test, preds))
```

```python
np.argmax(model.feature_importances_)
```

```python
X.columns[22]
```

```python
sns.countplot(data=data, x='odor', hue='class')
```

```python
error_rates = []

for n in range(1,96):
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    err = 1 - accuracy_score(y_test, preds)
    
    error_rates.append(err)
```

```python
plt.figure(figsize=(4,4),dpi=200)
plt.plot(range(1,96),error_rates)
```

```python
features = pd.DataFrame(index=X.columns, data=model.feature_importances_, columns=['Importance'])
```

```python
features
```

```python
imp_features = features[features['Importance']>0].sort_values(by='Importance')
```

```python
imp_features
```

```python
plt.figure(figsize=(8,5),dpi=200)
sns.barplot(data=imp_features, x=imp_features.index, y='Importance')
plt.xticks(rotation=90);
```

```python
#Gradient Boosting
```

```python
X1 = pd.get_dummies(data.drop('class', axis=1), drop_first=True)
y1 = data['class']
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=.1, random_state=101)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators':[50,100],
             "learning_rate":[0.1, 0.05, 0.2],
             "max_depth":[3,4,5]}
```

```python
gb_model = GradientBoostingClassifier()
grid = GridSearchCV(gb_model, param_grid)
grid.fit(X_train1, y_train1)
```

```python
grid.best_params_
```

```python
gb_preds = grid.predict(X_test1)
print(classification_report(y_test1, gb_preds))
```

```python
feature_imp = grid.best_estimator_.feature_importances_
```

```python
imp_features = pd.DataFrame(index=X1.columns,
                           data = feature_imp,
                           columns=['Importance'])
```

```python
imp_features
```

```python
imp_features = imp_features[imp_features['Importance']>0.0005].sort_values('Importance')
```

```python
imp_features
```

```python
plt.figure(figsize=(8,5),dpi=200)
sns.barplot(data=imp_features, x=imp_features.index, y='Importance')
plt.xticks(rotation=90);
```
