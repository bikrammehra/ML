```python
X = df.drop(columns=["sales", "total_spend"])
y = df["sales"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=.3, random_state=30)
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test  = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(scaled_X_train, y_train)
```

```python
linear_pred = linear_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE = mean_squared_error(y_test, linear_pred)
```

```python
MSE
```

```python
RMSE = np.sqrt(MSE)
```

```python
RMSE
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.scatterplot(y_test, range(0,len(y_test)))
sns.scatterplot(linear_pred, range(0,len(y_test)))
```

```python
residuals = y_test - linear_pred
```

```python
sns.scatterplot(x=y_test, y=residuals)
plt.axhline(y=0, color="red")
```

```python
sns.distplot(residuals,kde=True)
```

```python
#Fit a polynomial of degree 1
X = df["total_spend"]
y = df["sales"]
```

```python
poly_res = np.polyfit(X, y, 1)
```

```python
print(f"Slope : {poly_res[0]} \t Intercept : {poly_res[1]}")
```

```python
potential_spend = np.linspace(0,500, 100)
```

```python
potential_sales = poly_res[0] * potential_spend + poly_res[1]
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.scatterplot(x="total_spend", y="sales", data=df)
plt.plot(potential_spend, potential_sales, color="red")
```

```python
third_degree_poly = np.polyfit(X, y, 3)
```

```python
third_degree_poly
```

```python
potential_spend_third_degree = np.linspace(0,500, 100)
```

```python
potential_sales_third_degree = third_degree_poly[0] * potential_spend_third_degree ** 3 +\
                                third_degree_poly[1] * potential_spend_third_degree ** 2 +\
                                third_degree_poly[2] * potential_spend_third_degree + third_degree_poly[3]
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.scatterplot(x="total_spend", y="sales", data=df)
plt.plot(potential_spend_third_degree, potential_sales_third_degree, color="red")
```
