```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../Data/cancer_tumor_data_features.csv")
```

```python
data.head()
```

```python
# sns.heatmap(data, annot=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(data)
```

```python
scaled_X.mean(axis=0)
```

```python
scaled_X - scaled_X.mean(axis=0)
```

```python
covariance_matrix = np.cov(scaled_X, rowvar=False)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
eigen_values.shape
```

```python
eigen_vectors.shape
```

```python
n_components = 2
```

```python
sorted_key = np.argsort(eigen_values)[::-1][:n_components]
```

```python
sorted_key
```

```python
eigen_values, eigen_vectors = eigen_values[sorted_key], eigen_vectors[:,sorted_key]
```

```python
#Original data --> project --> eigen_vectors
principal_components = np.dot(scaled_X, eigen_vectors)
```

```python
principal_components
```

```python
plt.figure(figsize=(8,4), dpi=200)
plt.scatter(principal_components[:,0], principal_components[:,1])
```

```python
#Implementing PCA using Scikit-learn
```

```python
scikit_learn_pca_data = pd.read_csv("../Data/cancer_tumor_data_features.csv")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scikit_learn_scaler = StandardScaler()
```

```python
scaled_X1 = scikit_learn_scaler.fit_transform(scikit_learn_pca_data)
```

```python
scaled_X1
```

```python
from sklearn.decomposition import PCA
```

```python
help(PCA)
```

```python
pca_model = PCA(n_components=2)
```

```python
project_data_pca = pca_model.fit_transform(scaled_X1)
```

```python
plt.figure(figsize=(8,4), dpi=200)
plt.scatter(project_data_pca[:,0], project_data_pca[:,1], c=cancer_dict["target"], alpha=.5)
```

```python
pca_model.components_
```

```python
pca_data_comp = pd.DataFrame(data=pca_model.components_,columns=scikit_learn_pca_data.columns)
```

```python
plt.figure(figsize=(20,4), dpi=200)
sns.heatmap(pca_data_comp, annot=True)
```

```python
pca_model.explained_variance_ratio_
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_30 = PCA(n_components=30)
```

```python
pca_30_results = pca_30.fit(scaled_X1)
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_variance = []
```

```python
for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X1)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30), explained_variance)
plt.xlabel("Num of Components")
plt.ylabel("Variance Explained")
```
