```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
data = pd.read_csv("../Data/rock_density_xray.csv")
```

```python
data.head()
```

```python
data.columns = ['Signal', 'Density']
```

```python
sns.scatterplot(x=data['Signal'], y =data['Density'])
```

```python
X = data['Signal'].values.reshape(-1,1)
y = data['Density']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_preds = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test, lr_preds)
```

```python
np.sqrt(mean_squared_error(y_test, lr_preds))
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.scatterplot(x=data['Signal'], y =data['Density'])
signal_range = np.arange(0,100)
plt.plot(signal_range, lr_model.predict(signal_range.reshape(-1,1)))
```

```python
def run_model(model,X_train,y_train,X_test,y_test):
    
    # Fit Model
    model.fit(X_train,y_train)
    
    # Get Metrics
    
    preds = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test,preds))
    print(f'RMSE : {rmse}')
    
    # Plot results
    signal_range = np.arange(0,100)
    output = model.predict(signal_range.reshape(-1,1))
    
    
    plt.figure(figsize=(12,6),dpi=150)
    sns.scatterplot(x='Signal',y='Density',data=data,color='black')
    plt.plot(signal_range,output)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe = make_pipeline(PolynomialFeatures(2), LinearRegression())
```

```python
run_model(pipe, X_train, y_train, X_test, y_test)
```

```python
pipe = make_pipeline(PolynomialFeatures(10),LinearRegression())
run_model(pipe,X_train,y_train,X_test,y_test)
```
